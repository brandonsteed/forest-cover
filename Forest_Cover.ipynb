```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('covtype1.csv')
df
```

```python
df.columns
```

```python
df[['Hillshade_9am', 'Hillshade_Noon', 'Hillshade_3pm']]
```

```python
plt.figure(figsize=(10,6))

plt.hist(x=df['Hillshade_Noon'], color='blue', bins=[115, 125, 135, 145, 155, 165, 175, 185, 195, 205, 215, 225, 235, 245, 255])
plt.title('Shade Scores within Roosevelt National Forest at Noon')
plt.xlabel('Shade Scores (out of 255)')
plt.ylabel('Number of Points')
```

The histogram shows the number of points within each range of shade scores.  As expected, most points in the forest are highly shaded.

```python
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1)
plt.hist(df['Hillshade_9am'], color='orange')
plt.title('Shade Scores at 9 AM (out of 255)')
plt.ylabel('Number of Points')
plt.xlabel('Shade Score')

plt.subplot(1, 2, 2)
plt.hist(df['Hillshade_3pm'], color='purple')
plt.title('Shade Scores at 3 PM (out of 255)')
plt.ylabel('Number of Points')
plt.xlabel('Shade Score')
```

The histograms show a comparison in the shade scores within Roosevelt National Forest from 9 AM to 3 PM.  The data shows that different points in the forest are generally more shaded in the morning than in the afternoon.

```python
plt.figure(figsize=(14,9))
           
plt.scatter(
    x=df['Horizontal_Distance_To_Roadways'],
    y=df['Hillshade_Noon'],
    color='green',
    marker='x')
plt.title('Relationship Between Distance to Roadways and Shade Scores')
plt.xlabel('Distance to Roadway')
plt.ylabel('Shade Score')
```

The scatter plot show shade scores in relation to distance from the nearest roadway.  Points can be completely shaded any distance from a roadway.  However, geographic points further from roadways are less likely to have a low shade score or experience high levels of sunlight.

```python
spruce_fir = df[df['Cover_Type'] == 1]
aspen = df[df['Cover_Type'] == 5]
x = spruce_fir['Hillshade_Noon']
y = aspen['Hillshade_Noon']

plt.figure(figsize=(10,6))
plt.hist(x, density=True, bins=[95, 115, 135, 155, 175, 195, 215, 235, 255], color='blue', alpha=.5)
plt.hist(y, density=True, bins=[95, 115, 135, 155, 175, 195, 215, 235, 255], color='red', alpha=.5)
plt.xlabel('Shade Score (Out of 255)')
```

The green histogram shows the distribution of shade score where there is a spruce/fir cover type, while the yellow histogram shows the distribution of shade scores where there is an aspen cover type. Aspen forest is more likely to let in more sunlight than spruce/fir forest, as shown by the shade scores in the range up to 195.  Spruce/fir forest generally offers denser cover. However, points within the aspen cover type are slightly more likely to be within the highest range of shade scores.

```python
a = np.mean(df['Hillshade_9am'])
b = np.mean(df['Hillshade_Noon'])
c = np.mean(df['Hillshade_3pm'])

shade_change = np.array([a, b, c])
sc = pd.DataFrame(shade_change)
sc.columns = ['score']
sc.index = ['9_am', 'Noon', '3_pm']
sc
```

```python
plt.figure(figsize=(10,6))

plt.plot([212.14, 223.32, 142.53], color='gray')
plt.title('Change in Average Shade Score from 9 AM to 3 PM')
plt.ylabel('Shade Score (out of 255)')
cur_axes = plt.gca()
cur_axes.axes.get_xaxis().set_ticklabels([])
```

The amount of shade within the forest changes dramatically throughout the day.  From 9 AM to Noon, the amount of shade within the forest increases to a small degree.  From Noon to 3 PM, the amount of change decreases greatly.
